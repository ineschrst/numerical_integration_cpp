```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# making a plot of the Errors and Convergence
df = pd.read_csv("../data/errors_x_squared.csv")
```

```python
plt.loglog(df["n"], df["trapezoid"], label="Trapezoid")
plt.loglog(df["n"], df["simpson"], label="Simpson")
plt.loglog(df["n"], df["montecarlo"], label="Monte Carlo")
plt.legend()
plt.xlabel("n")
plt.ylabel("Error")
plt.show()
```
